```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import LabelEncoder, StandardScaler

from sklearn.linear_model import LinearRegression

from sklearn.tree import DecisionTreeRegressor

from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score

print('All libraries imported successfully!')
```

```python
df = pd.read_csv('student_performance.csv')
print(f'Dataset loaded: {df.shape[0]} students, {df.shape[1]} columns')
print(f'Columns: {df.columns.tolist()}')
print(f'\nMissing values: {df.isnull().sum().sum()}')
print('\nFirst 5 rows')
df.head()
```

```python
print('=== ML TASK DEFINITION ===')
print('TARGET : programming_score (the number we want to predict)')
print('FEATURE : math_score, science_score, english_score,attendance_percentage')
```

```python
df_ml = df.copy()
le_gender = LabelEncoder()
df_ml['gender_encoder'] = le_gender.fit_transform(df_ml['gender'])
print(f'Gender encoding: {dict(zip(le_gender.classes_,le_gender.transform(le_gender.classes_)))}')
le_dept = LabelEncoder()
df_ml['department_encoder'] = le_dept.fit_transform(df_ml['department'])
print(f'Department encoding: {dict(zip(le_dept.classes_,le_dept.transform(le_dept.classes_)))}')

print('\nNew columns added: gender_encoded, department_encoded')
df_ml[['gender','gender_encoder','department','department_encoder']].head(5)
```

```python
feature_cols = [
    'math_score',
    'science_score',
    'english_score',
    'attendance_percentage',
    'gender_encoder',
    'department_encoder'
]

x = df_ml[feature_cols]
y = df_ml['programming_score']

print(f'Feature matrix x shape: {x.shape} (students x features)')
print(f'Target vector y shape: {y.shape} (one score per student)')
print(f'\nFeature columns: {feature_cols}')
print(f'Target column: programming_score')
print(f'Target range: {y.min()} to {y.max()} (mean: {y.mean():.1f})')
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x,y,
    test_size=0.2,
    random_state=42
    )

print(f'Total students    : {len(x)}')
print(f'Training students : {len(x_train)} ({len(x_train) / len(x) * 100:.0f}%)')
print(f'Testing students  : {len(x_test)} ({len(x_test) / len(x) * 100:.0f}%)')
print(f'\nTraining target range: {y_train.mean()} - {y_train.max()}')
print(f'Testing target range : {y_test.min()} - {y_test.max()}')
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)


print('Features scaled successfully!')
print(f'Training feature mean (should be = 0): {x_train.mean(axis=0).mean():.4f}')
print(f'Training feature std (should be = 1): {x_train.std(axis=0).mean():.4f}')
print('\nNOTE : Scaler was fit on training data only - prevents data leakage')
```
