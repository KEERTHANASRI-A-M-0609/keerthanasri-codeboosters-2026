```python
raw_df = pd.read_csv('messy_sales_data.csv')

print(f'Raw data loaded: {raw_df.shape[0]} rows, {raw_df.shape[1]} columns')
print(f'Columns: {raw_df.columns.tolist()}')
print('\nFirst 5 rows:')
raw_df.head()
```

```python
print('=' * 55)
print(' DATA QUALITY DIAGNOSIS REPORT')
print('=' * 55)

#1.Missing values
print('\n[1] MISSING VALUES per column:')
print(raw_df.isnull().sum())

#2.Duplicates
print(f'\n[2] DUPLICATED ROWS:{raw_df.duplicated().sum()}')

#3.Data types
print('\n[3] DATA TYPES:')
print(raw_df.dtypes)

#4.Unique values in text columns(spot inconsistent)
print('\n[4] UNIQUE CATEGORIES:', raw_df['category'].unique())
print('[4] Sample customer name:', raw_df['customer_name'].dropna().unique()[:8])
print('[4] Sample order_date values:', raw_df['order_date'].unique()[:6])
```

```python
# Missing values for quantity column
print('\n[1] MISSING VALUES in quantity column:')
print(raw_df['quantity'].isnull().sum())
```

```python
#Create a working copy (ETL best practice)

df = raw_df.copy()
print(f'Working copy created: {df.shape}')
print('raw_df is untouched - we can always reset by running df = raw_df.copy()')
```

```python
#Fix #1: Handle Missing Values

print('Before fixing nulls:', df.isnull().sum().sum(), 'total missing values')

df['customer_name'].fillna('Unknown Customer', inplace = True)
median_qty = df['quantity'].median()
df['quantity'].fillna(median_qty, inplace = True)
print(f' Filled missing quantity with median: {median_qty}')

df['category'].fillna('Uncategorized', inplace = True)

print('After fixing nulls:',df.isnull().sum().sum(), 'total missing values')
```

```python
#Fix #2 : Remove Duplicates

print(f'Before deduplication: {len(df)} rows')
print(f'Duplicated rows: {df.duplicated().sum()}')
print('\nDuplicate rows:')
print(df[df.duplicated(keep=False)][['order_id','customer_name','quantity']])
df.drop_duplicates(inplace=True)
print(f'\nAfter deduplication: {len(df)} rows')
print(f'Rows removed: {len(raw_df) - len(df)}')
```

```python
# Fix #2 : Remove Rows with Missing Product
print(f'Before removing missing products: {len(df)} rows')
print(f'Missing product rows: {df["product"].isnull().sum()}')
print('\nRows with missing product:')
print(df[df['product'].isnull()][['order_id', 'customer_name', 'category']])

df.dropna(subset=['product'], inplace=True)
print(f'\nAfter removing missing products: {len(df)} rows')
print(f'Rows removed: {len(raw_df) - len(df)}')
```

```python
# Load the student dataset
student_df = pd.read_excel('student_performanceduple.csv.xlsx')

print(f'Before deduplication: {len(student_df)} rows')
print(f'Duplicated rows: {student_df.duplicated().sum()}')
print('\nDuplicate rows:')
# This will display all columns for the duplicated rows
print(student_df[student_df.duplicated(keep=False)])

# Save initial length to track removed rows
initial_rows = len(student_df)

# Remove duplicates
student_df.drop_duplicates(inplace=True)
print(f'\nAfter deduplication: {len(student_df)} rows')
print(f'Rows removed: {initial_rows - len(student_df)}')
```

```python
print('Sample dates before parsing:')
print(df['order_date'].head(8).tolist())

# Convert to datetime object first (keeping dayfirst=True)
df['order_date'] = pd.to_datetime(
    df['order_date'],
    dayfirst=True,
    format='mixed',
    errors='coerce'
)

nat_count = df['order_date'].isnull().sum()
print(f'\nUnparseable dates (NaT): {nat_count}')

# Extract year, month, and month name
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
df['month_name'] = df['order_date'].dt.strftime('%B')

# Format the order_date column specifically to Day-Month-Year for display
df['order_date'] = df['order_date'].dt.strftime('%d-%m-%Y')

print('\nSample dates after parsing:')
print(df[['order_date', 'year', 'month', 'month_name']].head(5))
```

```python
print('Before Standardization:', df['customer_name'].unique()[:6])

df['customer_name'] = (
    df['customer_name']
    .str.strip()
    .str.title()
)

print('After Standardization:', df['customer_name'].unique()[:6])
wrong_mask = df['product'] == 'Keyboard'
df.loc[wrong_mask, 'category'] = 'Accessories'

print('After fix: Unique categories:', df['category'].unique())
```

```python
#Fix Data Types + Create Revenue Column

df['revenue'] = df['quantity'] * df['unit_price']
print('Revenue column created:')

print(df[['customer_name', 'product', 'quantity', 'unit_price', 'revenue']].head(5))

# Calculate the total revenue across all orders
total_revenue = df['revenue'].sum()
print('\n')
print(f"Total revenue across all ordwers: rupees{int(total_revenue)}")
```
