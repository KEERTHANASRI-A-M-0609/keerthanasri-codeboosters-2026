```python
import pandas as pd
print("Libraries imported sucessfully!")
```

```python
df = pd.read_csv('student_performance.csv')
```

```python
print(df)
```

```python
print("==============================")
print("Student Performance Dataset")
print("="*30)
print("\n")
print("[1] DATASET OVERVIEW")
print("-"*80)

print("Total Students: ",df.shape[0])
print("Total Columns: ",df.shape[1])
print("Departments: ",df['department'].unique())
print("No. of departments: ",df['department'].nunique())
print("\n")

print("[2] DEPARTMENT-WISE STUDENT COUNT")
print("-"*80)
print(df['department'].value_counts())
print("\n")

print("[3] SCORE ANALYSIS")
print("-"*80)
print("Highest Math score: ",df['math_score'].max())
print("Lowest Math score: ",df['math_score'].min())
print(f"Average Math score: {df['math_score'].mean():.2f}")
print(f"Average Programming score: {df['programming_score'].mean():.2f}")
print(f"Average Attendance: {df['attendance_percentage'].mean():.2f}%")
print("\n")

print("[4] GENDER-WISE AVERAGE SCORES")
print("-"*80)
df['total_score'] = df['math_score'] + df['science_score'] + df['english_score'] + df['programming_score']
gender_scores = df.groupby('gender')['total_score'].mean().round(2)
print(gender_scores)
print("\n")

print("[5] TOP 5 PERFORMERS")
print("-"*80)
top5 = df.sort_values(by='total_score', ascending=False).head(5)
for index,row in top5.iterrows():
    print(f"{row['name']:<20} | {row['department']:<18} | Score: {row['total_score']}")
print("\n")

print("[6] STUDENTS WITH LOW ATTENDANCE(Below 75%)")
print("-"*80)
low_att = df[df['attendance_percentage'] < 75][['name','department','attendance_percentage']]
if len(low_att) > 0:
  print(low_att.to_string(index=False))
else:
  print("All students have attendance above 75%.")

print("\n" + "="*80)
print("Report Complete")
print("="*80)
print("\n")
```
